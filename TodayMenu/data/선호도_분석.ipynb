```python
import pandas as pd

# 데이터 로드
data = pd.read_csv('./FOOD_POR_RANK_.csv')
print(data.head())
```

```python
# 결측치 처리
data = data.dropna()

# 데이터 정규화 (선호도 점수는 0에서 1 사이로 스케일링)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data[['SE_ID']] = scaler.fit_transform(data[['SE_ID']])
```

```python
# 사용자별 선호도 평균
user_avg_preference = data.groupby('SE_ID')['POR_RANK'].mean()

# 음식 카테고리별 선호도 평균
category_avg_preference = data.groupby('FOOD_NM')['POR_RANK'].mean()
```
